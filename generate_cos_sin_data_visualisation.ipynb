```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd

def plot_data(filepath, start_time=0, end_time=500):
    """
    Plots a subset of data from the given filepath.
    
    Args:
    - filepath (str): Path to the CSV file.
    - start_time (float): Start time for the subset of data to plot.
    - end_time (float): End time for the subset of data to plot.
    """
    
    # Load data from the file
    data = np.loadtxt(filepath, delimiter=',')
    sin_data = data[:, 0]
    cos_data = data[:, 1]
    
    sampling_rate = 60  # 500 Hz, chosen as a moderate sampling rate
    time_duration = 1000 # seconds
    time_points = np.linspace(0, time_duration, int(sampling_rate * time_duration))
    
    # Get indices for the desired time range
    start_idx = int(start_time * sampling_rate)
    end_idx = int(end_time * sampling_rate)
    
    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(time_points[start_idx:end_idx], sin_data[start_idx:end_idx], label='Sin', color='blue')
    plt.plot(time_points[start_idx:end_idx], cos_data[start_idx:end_idx], label='Cos', color='red')
    plt.title(f"Visualization of {os.path.basename(filepath)}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example usage:
plot_data("generated_files/P01_0.5sincos0.25pi.csv", start_time=0, end_time=8)
```
